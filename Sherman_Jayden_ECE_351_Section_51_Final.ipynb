### Question 1 (10 points)

```python
# Insert code for Question 1 here
steps = 1e-3

plt.figure(figsize=(12,12))
plt.suptitle("Question 1")


t = np.arange(0, 3 + steps, steps)
y = 4*np.cos(t)

plt.subplot(4, 1, 1)
plt.plot(t,y,color = "red")
plt.grid()
plt.ylabel("$x_1$")
plt.xlabel("t [s]")
plt.legend(["$4cos(t)$"], loc = "upper right")


t = np.arange(0, 2 + steps, steps)
y = 1.5*np.sin(2*t)

plt.subplot(4, 1, 2)
plt.plot(t,y, color = "blue")
plt.grid()
plt.ylabel("$x_2$")
plt.xlabel("t [s]")
plt.legend(["$1.5sin(2t)$"], loc = "upper right")


t = np.arange(0, 5 + steps, steps)
y = 3*(np.cos(3*t)**2)*np.exp(-t)

plt.subplot(4, 1, 3)
plt.plot(t,y, color = "green")
plt.grid()
plt.ylabel("$x_3$")
plt.xlabel("t [s]")
plt.legend(["$3cos^2(3t)e^{-t}$"], loc = "upper right")


t = np.arange(0, 6 + steps, steps)
y = 0.5*np.exp(-5*t)

plt.subplot(4, 1, 4)
plt.plot(t,y, color = "black")
plt.grid()
plt.ylabel("$x_4$")
plt.xlabel("t [s]")
plt.legend(["$0.5e^{-5t}$"], loc = "upper right")


plt.tight_layout()
plt.show()
```

```python
# Insert code for Question 2 Part 1 here
steps = 1e-4

plt.figure(figsize = (12,6))
plt.suptitle("Question 2 - Part 1")


t = np.arange(0, 3 + steps, steps)
y = 4*np.cos(t)

plt.subplot(2, 2, 1)
plt.plot(t,y, color = "black")
plt.grid()
plt.xlabel("t [s]")
plt.title("$x_1$")


t = np.arange(0, 2.5 + steps, steps)
t_len = len(t)
t_ext = np.arange(0, 2 * t[t_len - 1], steps)

f1 = 4*np.cos(t)
f2 = 1.5*np.sin(2*t) 
conv_1_2 = sig.convolve(f1,f2)*steps

plt.subplot(1, 2, 2)
plt.plot(t_ext,conv_1_2[:len(t_ext)], color = "red")
plt.grid()
plt.xlabel("t [s]")
plt.title("$x_1*x_2$")


t = np.arange(0, 2 + steps, steps)
y = 1.5*np.sin(2*t)

plt.subplot(2, 2, 3)
plt.plot(t,y, color = "black")
plt.grid()
plt.xlabel("t [s]")
plt.title("$x_2$")


plt.tight_layout()
plt.show()
```
