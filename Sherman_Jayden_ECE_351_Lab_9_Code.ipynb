```python
#Insert all necessary imports here

import scipy.signal as sig
import scipy.fftpack as sfft
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Insert your function to perform the fft here

def my_fft(x, f_s):
    """
    Perform a Fast Fourier Transform
    Inputs: x - function, f_s - sampling frequency
    Oututs: freq - frequencies for output, X_mag - magnitudes of signal, X_phi - phase of signal
    """
    N = len(x) #find the length of the signal
    X_fft = sfft.fft(x) #perform the fft
    X_fft_shifted = sfft.fftshift(X_fft) #shift zero frequency components to the center of the spectrum
    freq = np.arange(-N/2, N/2) * f_s / N #compute the frequencies for the output signal. fs is the sampling frequency
    X_mag = np.abs(X_fft_shifted) / N #compute the magnitudes of the signal
    X_phi = np.angle(X_fft_shifted) #compute the phase of the signal
    return freq, X_mag, X_phi
```

```python
#Insert code for Part 1 Task 1 here

#Set up parameters
f_s = 100
steps = 1/f_s
t = np.arange(0, 2, steps)

#Define function
x = np.cos(2*np.pi*t)

#Initiate plot
plt.figure()
plt.suptitle("Task 1 - User-Defined FFT of cos(2$\pi$t)")

#Time domain plot
plt.subplot(3, 1, 1)
plt.plot(t,x)
plt.grid()
plt.ylabel("x(t)")
plt.xlabel("t [s]")

#Take Fast Fourier Transform
freq, X_mag, X_phi = my_fft(x, f_s)

#Plot magnitude
plt.subplot(3, 2, 3)
plt.stem(freq, X_mag)
plt.grid()
plt.ylabel("|X(f)|")

plt.subplot(3, 2, 4)
plt.stem(freq, X_mag)
plt.xlim(-2, 2)
plt.grid()

#Plot Phase
plt.subplot(3, 2, 5)
plt.stem(freq, X_phi)
plt.grid()
plt.ylabel("/_X(f)")
plt.xlabel("f [Hz]")

plt.subplot(3, 2, 6)
plt.stem(freq, X_phi)
plt.xlim(-2, 2)
plt.xlabel("f [Hz]")
plt.grid()

#Finish Plot
plt.tight_layout()
plt.show()
```
