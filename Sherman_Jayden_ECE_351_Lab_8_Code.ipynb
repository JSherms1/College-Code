Consider the square wave in the figure. Assume that $x(t)$ is a real-value function and can be represented by the Fourier series:  
$$x(t) = \frac{1}{2} a_0 + \displaystyle \sum_{k =1}^{\infty}a_kcos(k\omega_0t) + b_ksin(k\omega_0t)$$
Where,  
$$a_k = \frac{2}{T}\displaystyle \int_0^Tx(t)cos(k\omega_0t)dt,$$  
$$b_k = \frac{2}{T}\displaystyle \int_0^Tx(t)sin(k\omega_0t)dt,$$
and  
$$\omega_0 = \frac{2\pi}{T}.$$

2. For the square wave in the figure, plot the Fourier series approximation for $N = \left\{1, 3, 15, 50, 150, 1500\right\}$ for,  
$$x(t) = \frac{1}{2} a_0 + \displaystyle \sum_{k =1}^{N}a_kcos(k\omega_0t) + b_ksin(k\omega_0t).$$  
Set $T = 8$ s and plot the approximation time for $0\le t \le 20$ s using an appropriate step size. Format the plots into 2 figures, each having 3 vertical subplots.

```python
#Insert all code for Part 1, Task 2 here
def x(t,N):
    """
    Calculate x(t) with t
    Inputs: t - time
    Oututs: x(t) - resulting function
    """ 
    sum = 0
    for k in range(1, N + 1):
        sum += a_k(k)*np.cos(k*w_0*t) + b_k(k)*np.sin(k*w_0*t)
    
    return (0.5*a_0) + sum

T = 8
w_0 = (2*np.pi)/T

steps = 1e-3
t = np.arange(0, 20 + steps, steps)


plt.figure()
plt.suptitle("Fourier Series Approximation")

y = x(t,1)

plt.subplot(3, 1, 1)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=1")
plt.xlabel("t [s]")

y = x(t,3)

plt.subplot(3, 1, 2)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=3")
plt.xlabel("t [s]")

y = x(t,15)

plt.subplot(3, 1, 3)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=15")
plt.xlabel("t [s]")

plt.tight_layout()
plt.show()


plt.figure()

y = x(t,50)

plt.subplot(3, 1, 1)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=50")
plt.xlabel("t [s]")

y = x(t,150)

plt.subplot(3, 1, 2)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=150")
plt.xlabel("t [s]")

y = x(t,1500)

plt.subplot(3, 1, 3)
plt.plot(t,y)
plt.grid()
plt.ylabel("x(t) N=1500")
plt.xlabel("t [s]")

plt.tight_layout()
plt.show()
```
